```python
import pandas as pd
import numpy as np
df_hotels = pd.read_json('/Users/fezzibasma/Desktop/Data/Data_Full-Stack/Data Collection & Management/ProjetBooking/I) data_extract/src/hoteldatabooking_bestpercitycenter.json',orient='records')
df_hotels.head(425)
```

```python
df_hotels['description'] = df_hotels['description'].apply(lambda x : (x[:143] + '..') if len(x)>143 
                                                                                      else x)
```

```python
df_hotels['rating']=df_hotels['rating'].apply(lambda x: str(x.replace(',','.')))
df_hotels
```

```python
data_location = df_hotels['location'].str.split(",",n=0,expand=True)
data_location['location_splitted'] = data_location[1].fillna(data_location[0])
data_location=data_location['location_splitted']
```

```python
df_hotels=pd.concat([df_hotels,data_location],join='outer',axis=1)
#concat 
```

```python
df_hotels['location_splitted']=df_hotels['location_splitted'].str.lstrip() #sert à supprimer les espaces en début de la chaîne de caractères
```

```python
#afficher les colonnes de location splitted
#print(df['location_splitted'].unique())
#reformater le mont saint michel au meme format de la liste en entrée du projet : 
df_hotels.location_splitted[df_hotels.location_splitted == "Le Mont-Saint-Michel"] = "Mont Saint Michel"
df_hotels.location_splitted[df_hotels.location_splitted == "Saint Malo"] = "St Malo"

```

```python
df_hotels['city'] = df_hotels['location_splitted'].apply(lambda x : 'Chateau du Haut Koenigsbourg' if (x == 'Saint-Hippolyte') or (x== 'Château du Haut-Kœnigsbourg')
                                                      else 'St Malo' if x=='Saint-Malo'
                                                      else 'Besancon' if x=='Besançon'
                                                      else 'Uzes' if x=='Uzès'
                                                      else 'Nimes' if x=='Nîmes'
                                                      else 'Saintes Maries de la mer' if x=='Les Saintes-Maries-de-la-Mer'
                                                      else 'Aix en Provence' if x=='Aix-en-Provence'
                                                      else 'Aigues Mortes' if x=='Aigues-Mortes'
                                                      else 'Gorges du Verdon' if (x == 'Moissac-Bellevue') or (x == 'Gréoux-les-Bains') 
                                                                                                           or (x == 'Moustiers-Sainte-Marie') or (x == 'Quinson')
                                                                                                           or (x == 'Trigance')or (x == 'Saint-Laurent-du-Verdon')
                                                                                                           or (x == 'Castellane') or (x == 'La Palud-sur-Verdon')
                                                                                                           or (x == 'La Palud-sur-Verdon')
                                                      else 'Annecy' if x == 'Sévrier'
                                                      else 'Bormes les Mimosas' if (x == 'Le Lavandou') or (x=='Bormes-les-Mimosas')
                                                      else 'Avignon' if x == 'Villeneuve-lès-Avignon'
                                                      else 'Eguisheim' if x == 'Husseren-les-Châteaux'
                                                      else 'Collioure' if x == 'Argelès-sur-Mer'
                                                      else 'Collioure' if x == 'Port-Vendres'
                                                      else 'Grenoble' if x == 'Saint-Martin-dʼHères'
                                                      else 'Bayonne' if x == 'Anglet'
                                                      else 'Ariege' if (x == 'Varilhes') or (x == 'Tarascon-sur-Ariège') 
                                                                                         or (x == 'Mercus-Garrabet') or (x == 'Mirepoix')
                                                                                         or (x == 'Trigance') or (x == 'Saint-Girons')
                                                                                         or (x == 'Bélesta') or (x == 'Le Bosc')
                                                                                         or (x == 'Foix') or (x =='Ax-les-Thermes') or(x=='Ariège')
                                                                                         
                                                     else x
                                                      )
```
